```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
```

# Preprocessing

```python
#taking only text which are alphabets
#iterating through all the sms data and removing characters other than the alphabets

dataset['cleaned'] = dataset['Review'].apply(lambda x:re.sub('[^A-Za-z]',' ',x))
```

# Text Normalization

Making all text lower case,this will be easy when we come to word vectorization

```python
dataset['lower'] = dataset['cleaned'].apply(lambda x:x.lower())
```
